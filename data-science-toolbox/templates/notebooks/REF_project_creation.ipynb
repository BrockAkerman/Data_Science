# 🗂️ DS Toolkit — README & Usage Guide

## What this is
A reusable, copy-paste-ready collection of data science pipeline components.  
Each notebook is a standalone, generic template you can drop into any project.

---

## 📚 Notebooks in this Toolkit

| # | Notebook | What it does |
|---|----------|-------------|
| 01 | `01_imports_and_setup.ipynb` | Standard imports, settings, data loading, preliminary inspection |
| 02 | `02_data_cleaning_pipeline.ipynb` | Duplicates, missing values, data types, outliers, value correction — all with audit trails |
| 03 | `03_eda.ipynb` | Target distribution, histograms, boxplots, default-rate-by-feature, correlation heatmap |
| 04 | `04_train_val_test_split.ipynb` | Stratified 60/20/20 split, class imbalance audit, SMOTE option, leakage check |
| 05 | `05_model_training.ipynb` | Logistic Regression, Decision Tree, Random Forest, Gradient Boosting, LightGBM, XGBoost + RandomizedSearchCV tuning |
| 06 | `06_model_evaluation.ipynb` | AUC-ROC, Gini, KS, Precision-Recall, Brier Score, Gains/Lift, Success Scorecard |
| 07 | `07_interpretability_shap.ipynb` | SHAP global/beeswarm/waterfall, permutation importance, PDPs, triangulation table |

---

## 🚀 How to Start a New Project

```
1. Copy all notebooks into your new project folder.
2. Open 01_imports_and_setup.ipynb and update DATA_PATH.
3. Open 02_data_cleaning_pipeline.ipynb and update the CONFIG dicts.
4. Work through notebooks 03 → 07 in order.
5. Every cell that needs changing is marked with # CHANGE
```

---

## ✏️ What to Change in Each Notebook

### 01 — Imports & Setup
- `DATA_PATH` → path to your CSV

### 02 — Data Cleaning
- `DUPLICATE_CONFIG['customer_id_column']` → your unique ID column
- `MISSING_VALUE_CONFIG['features']` → add entries for your columns with nulls
- `DATA_TYPE_CONFIG` → your numeric and categorical column lists
- `OUTLIER_CONFIG['features']` → your numeric features and treatment strategies

### 03 — EDA
- `TARGET_COL` → your target column name
- Feature lists in plot calls

### 04 — Split
- `TARGET` → your target column
- `DROP_COLS` → ID columns, EDA artifacts, raw target strings

### 05 — Model Training
- Hyperparameter values in each model block (annotated with # CHANGE)
- `APPLY_SMOTE = True` if your minority class < 10%

### 06 — Evaluation
- Threshold values in the Success Scorecard

### 07 — Interpretability
- `OPTIMAL_THRESHOLD` → from your evaluation notebook
- `SHAP_SAMPLE` → increase for more stable estimates

## 💡 Tips

- Always run 02 (cleaning) before 03 (EDA) — never EDA on raw data.
- Never look at the test set until you're ready for final evaluation (notebook 06).
- The `model_registry` dict in notebook 05 persists all models for fair comparison.
- SHAP TreeExplainer only works for tree-based models — use KernelExplainer for LR.
- Permutation importance is the most reliable model-agnostic method.

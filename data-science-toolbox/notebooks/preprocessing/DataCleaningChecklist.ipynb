### Dealing with Duplicates

##### locate and look

```python
# Locate Duplicates
data.duplicated().value_counts()
```

```python
# Inspect some rows containing duplicates as needed
data[data.duplicated(keep=False)].sort_values(list(data.columns))
```

```python
# Check before removing
print(f"Duplicates found: {data.duplicated().sum()}")
```

##### decisioning on duplicates

```python
# Remove all Duplicates
data = data.drop_duplicates()
```

```python
# Duplicate removal based on columns
data= data.drop_duplicates(subset=[''], keep='first')
data= data.drop_duplicates(subset=['',''])
```

### Missing Values

```python
# Strategy #1: Fill with meaningful defaults
data[''] = data[''].fillna()
data[''] = data[''].fillna('some_comment')
```

```python
# Strategy #2:  Drop rows with critical missing data
data = data.drop_na(subset=['',''])
```

```python
# Strategy #3:  Fill numerical with mean/median
data[''] = data[''].fillna(data[''].median())
data[''] = data[''].fillna(data[''].mean())
```

```python
# Strategy #4:  Fill categorical with mode
most_common = data[''].mode()[0]
data[''] = data[''].fillna(most_common)
```

```python
# STrategy #5:  Forward/Backward fill for time-series
data[''] = data[''].fillna(method='ffill')
```

### Standardize Text

```python
# Fix inconsistencies
data['<email>'] = data['<email>'].str.lower()
data['<state>'] = data['<state>'].str.upper()
data['<city>'] = data['<city>'].str.title()
```

```python
# Remove extra whitespaces
data[''] = data[''].str.strip()
data[''] = data[''].str.replace(r'\s+', ' ', regex=True)

data['status'] = data['status'].replace({
    'Y': 'Yes', 'N':'No', 'T':'True', 'F':'False', 'active':'Active', 'ACTIVE':'Active',
})
```

### Fix Datatypes

```python
# Convert to Correct Types

# Arguments for astype() method include 'string', 'int', 'float', 'bool'
data[''] = data[''].astype()
```

```python
# Parse Dataes Properly
pd.to_datetime(data['<time>'], format='%Y-%m-%d')
pd.to_datetime(data['<time>'], errors='coerce')
```

```python
# Extract Data Components
data['year'].dt.year
data['month'].dt.month
data['day_of_week'].dt.day_name()
```

### Remove Invalid Dates

```python
# Filter out impossible values
data[data['age'] > 0]
data[data['price'] >= 0]
data[data['rating'].between(1,5)]
data[data['quantity'] > 0]
```

```python
# Drop columns with excessive missing data
threshold = 0.5
missing_pct = data.isnull().mean()
cols_to_drop = missing_pct[missing_pct > threshold].index
data = data.drop(columns=cols_to_drop)
```

### Outliers

```python
# Interquartile Range Method
Q1 = data[''].quantile(0.25)
Q3 = data[''].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Alternative numpy version
# Q1_df = data[''].quantile(q=[0.25, 0.50, 0.75])
```

```python
data = data[(data[''] >= lower_bound) & (data[''] <= upper_bound)]
```

```python
# Winsorization Method caps outliers

data[''] = data[''].clip(lower=lower_bound, upper=upper_bound)
```

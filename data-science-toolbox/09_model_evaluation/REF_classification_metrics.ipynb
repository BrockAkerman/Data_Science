```python
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve, precision_recall_curve,
                             average_precision_score)
import matplotlib.pyplot as plt, seaborn as sns, numpy as np, pandas as pd

# ── Full Report ───────────────────────────────────────────────────────────
# y_pred = model.predict(X_val)
# y_prob = model.predict_proba(X_val)[:, 1]
# print(classification_report(y_val, y_pred))
# print(f'ROC-AUC: {roc_auc_score(y_val, y_prob):.4f}')
```

```python
# ── Confusion Matrix ─────────────────────────────────────────────────────
def plot_confusion_matrix(y_true, y_pred, labels=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=labels, yticklabels=labels)
    plt.ylabel('Actual'); plt.xlabel('Predicted')
    plt.title('Confusion Matrix')
    plt.tight_layout(); plt.show()

# plot_confusion_matrix(y_val, y_pred, labels=['No', 'Yes'])
```

```python
# ── ROC Curve ─────────────────────────────────────────────────────────────
def plot_roc_curve(y_true, y_prob, model_name='Model'):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    auc = roc_auc_score(y_true, y_prob)
    plt.figure(figsize=(7, 5))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC={auc:.3f})', lw=2)
    plt.plot([0,1],[0,1], 'k--', lw=1)
    plt.xlabel('FPR'); plt.ylabel('TPR')
    plt.title('ROC Curve'); plt.legend()
    plt.tight_layout(); plt.show()

# plot_roc_curve(y_val, y_prob)
```

```python
# ── Precision-Recall Curve ────────────────────────────────────────────────
def plot_pr_curve(y_true, y_prob, model_name='Model'):
    prec, rec, _ = precision_recall_curve(y_true, y_prob)
    ap = average_precision_score(y_true, y_prob)
    plt.figure(figsize=(7, 5))
    plt.plot(rec, prec, label=f'{model_name} (AP={ap:.3f})', lw=2)
    plt.xlabel('Recall'); plt.ylabel('Precision')
    plt.title('Precision-Recall Curve'); plt.legend()
    plt.tight_layout(); plt.show()

# Use PR curve when classes are imbalanced — it's more informative than ROC
# plot_pr_curve(y_val, y_prob)
```

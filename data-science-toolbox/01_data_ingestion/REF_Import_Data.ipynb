## 📁 Load Data

```python
import pandas as pd
import io
import requests

def load_data(source, file_type='csv', sheet_name=0, json_orient='records'):
    """
    Standardizes data ingestion into a single 'df' object.
    Supports: CSV, Excel, JSON, URLs, and Google Drive.
    """
    try:
        # 1. Handle Google Drive "Share" Links
        if "drive.google.com" in source:
            file_id = source.split('/')[-2] if '/d/' in source else source.split('id=')[-1]
            source = f'https://drive.google.com/uc?export=download&id={file_id}'
        
        # 2. Ingest based on file type
        if file_type == 'csv':
            df = pd.read_csv(source)
        elif file_type == 'excel':
            df = pd.read_excel(source, sheet_name=sheet_name)
        elif file_type == 'json':
            df = pd.read_json(source, orient=json_orient)
        elif file_type == 'parquet':
            df = pd.read_parquet(source)
        else:
            raise ValueError("Unsupported file type. Use 'csv', 'excel', 'json', or 'parquet'.")
            
        print(f"✅ Successfully loaded data from: {source[:50]}...")
        return df

    except Exception as e:
        print(f"❌ Error loading data: {e}")
        return None

# --- Quick Reference Usage ---

# Local CSV
# df = load_data('data_v1.csv')

# Excel with specific sheet
# df = load_data('annual_report.xlsx', file_type='excel', sheet_name='Q4_Stats')

# Public GitHub/Web Link
# df = load_data('https://raw.githubusercontent.com/user/repo/main/data.csv')

# Google Drive (using the shared 'Copy Link' URL)
# df = load_data('https://drive.google.com/file/d/1XyZ.../view?usp=sharing')
```
